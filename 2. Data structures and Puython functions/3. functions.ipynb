### 1. *args and **kwargs

```python
# concepts of *args

def multiply(*args):
    # initialize product to 1 (neutral value for multiplication)
    product = 1  

    # iterate through all arguments passed
    for i in args:
        # multiply each value with current product
        product = product * i  
    
    # show all input values (tuple of arguments)
    print(args)  
    
    # return final multiplication result
    return product  

multiply(1, 2, 3, 4, 5, 6)
```

```python
# Concept of **kwargs

def display(**kwargs):
    # iterate through key-value pairs in kwargs dictionary
    for (keys, values) in kwargs.items():
        # print each key and its corresponding value
        print(keys, '->', values)

# pass multiple keyword arguments (stored as a dictionary)
display(Nepal="Kathmandu", India="Delhi", Sirlanka="Colombu")
```

### Concept of local and global variable

```python
# Example 1

def g(y):
    # access global variable x and print it
    print(x)
    
    # use global x to compute and print x + 1
    print(x + 1)

# define global variable x
x = 5

# call function g and pass x (argument y is not used inside)
g(x)

# print global x again (unchanged)
print(x)
```

```python
# Example 2

def f(y):
    # create a local variable x inside the function
    x = 1
    
    # increment local x by 1
    x += 1
    
    # print the local value of x
    print(x)

# define global variable x
x = 5

# call function (y receives 5, but it's not used inside)
f(x)

# print global x (remains unchanged)
print(x)
```

```python
# Example 3


# def f(y):
#     # this would cause an error because x is treated as local but not defined
#     x += 1

# x = 5
# f(x)
# print(x)


def h(y):
    # declare that we want to use the global variable x
    global x
    
    # increment the global x by 1
    x += 1

# define global variable x
x = 5

# call function (y is passed but not used)
h(x)

# print updated global x
print(x)
```

### Concept of nested loops

```python
def f():
    # define inner function g inside f
    def g():
        # print message from inner function
        print("inside function g")
    
    # call the inner function g
    g()

    # print message from outer function f
    print("Inside functionn f")

# call outer function f
f()
```

### Higher order functions

```python
#example

# normal function → takes value and returns result
def square(x):
    return x**2

# higher-order function → takes another function as argument
def transform(f, l):
    # store results
    output = []

    # apply function f to each element
    for i in l:
        output.append(f(i))

    return output

l = [1, 2, 3, 4, 5]

# passing normal function to higher-order function
print(transform(square, l))

# passing lambda (anonymous function) to higher-order function
print(transform(lambda x: x**3, l))
```

### Examples of higher order functions

1. map()

```python
#square the items of the list

# use map (higher-order function) to apply lambda (normal function) on each element
list(map(lambda x: x * 2, [1, 2, 3, 4, 5]))
```

```python
#odd even labeling of list items

# use map (higher-order function) with lambda (normal function)
# to label each number as 'even' or 'odd'
list(map(lambda x: 'even' if x % 2 == 0 else 'odd', [1, 2, 3, 4, 5]))
```

```python
# fetch names form a list of dictionary

# list of dictionaries (user data)
users = [
    {
        'name': 'Rahul',
        'age': 45,
        'gender': 'male'
    },
    {
        'name': 'Nitish',
        'age': 33,
        'gender': 'male'
    },
    {
        'name': 'Ankita',
        'age': 50,
        'gender': 'female'
    }
]

# use map (higher-order function) to extract 'name' from each dictionary
# lambda takes each dictionary and returns the value of 'name'
list(map(lambda user: user['name'], users))
```

### 2. filter()

```python
#numbers greater then 5

# use filter (higher-order function) with lambda (normal function)
# to keep only numbers greater than 5
list(filter(lambda x: x > 5, [2, 3, 5, 6, 4, 7, 1, 8, 9]))
```

```python
#fetch fruits starting with a

# use filter (higher-order function) with lambda (normal function)
# to select only fruits that start with letter 'a'
list(filter(lambda fruit: fruit.startswith('a'), ['apple', 'mango', 'guava']))
```

### 3. reduce()

```python
#sum of all items

import functools

# use reduce (higher-order function) to cumulatively add elements of the list
# lambda takes two values at a time and returns their sum
functools.reduce(lambda x, y: x + y, [1, 2, 3, 4, 5])
```

```python
# find max

import functools

# use reduce (higher-order function) to find the maximum value in the list
# lambda compares two values and keeps the larger one
functools.reduce(lambda x, y: x if x > y else y, [1, 2, 3, 4, 5])
```
