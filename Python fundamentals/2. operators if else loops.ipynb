### `Problem 6` - Find the factorial of a given number.

Write a program to use the loop to find the factorial of a given number.

The factorial (symbol: `!`) means to multiply all whole numbers from the chosen number down to 1.

For example: calculate the factorial of 5

```bash
5! = 5 × 4 × 3 × 2 × 1 = 120
```

Output:

```bash
120
```

````python
# Take input from user
num = int(input("Enter the number: "))

# Initialize factorial as 1
fact = 1

# Loop from 1 to num
for i in range(1, num + 1):
    fact = fact * i   # multiply each number

# Print result
print("The factorial of", num, "is", fact)
````

### `Problem 7` - Reverse a given integer number.

Example:

`Input:`

```bash
76542
```

`Output:`

```bash
24567
```

````python
# Take input from user
num1 = int(input("Enter the number: "))

# Store reversed number
reverse = 0

# Loop until number becomes 0
while num1 != 0:
    lsb = num1 % 10          # get last digit
    reverse = reverse * 10 + lsb   # add digit to reverse
    num1 = num1 // 10        # remove last digit

# Print reversed number
print(reverse)
````

### `Problem 8`: Take a user input as integer N. Find out the sum from 1 to N. If any number if divisible by 5, then skip that number. And if the sum is greater than 300, don't need to calculate the sum further more. Print the final result. And don't use for loop to solve this problem.

**Example 1:**

`Input:`

```bash
30
```

`Output:`

```bash
276
```

````python
# Take input
n = int(input('enter the number: '))

# Initialize sum
sum = 0

# Start from 1
i = 1

# Loop from 1 to n
while i < n + 1:

    # Skip multiples of 5
    if i % 5 == 0:
        i += 1
        continue

    # Add number to sum
    sum += i

    # If sum exceeds 300, undo last addition and stop
    if sum > 300:
        sum -= i
        break

    # Move to next number
    i += 1

# Print final sum
print(sum)
````

````python
# Loop from 1000 to 3000
for i in range(1000, 3001):

    flag = True      # assume number is valid
    num = i          # copy of number

    # Check each digit
    while num > 0:
        lsb = num % 10   # get last digit

        # if digit is odd → reject number
        if lsb % 2 != 0:
            flag = False
            break

        num = num // 10  # remove last digit

    # if all digits are even → print number
    if flag:
        print(i, end=' ')
````

### `Problem 12`: A robot moves in a plane starting from the original point (0,0). The robot can move toward UP, DOWN, LEFT and RIGHT with a given steps.
The trace of robot movement is shown as the following:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
> The numbers after the direction are steps.

> `!` means robot stop there.

**Please write a program to compute the distance from current position after a sequence of movement and original point.**

*If the distance is a float, then just print the nearest integer.*

Example:

`Input`:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
`Output`:
```
2
```

````python
# Starting position (x, y)
pos = [0, 0]

while True:
    s = input("Enter the robot path: ")

    # Stop condition
    if s == '!':
        break

    # Split input into direction and steps
    direction = s.split()[0]
    steps = int(s.split()[1])

    # Move robot based on direction
    if direction == 'UP':
        pos[1] += steps

    elif direction == 'DOWN':
        pos[1] -= steps

    elif direction == 'RIGHT':
        pos[0] += steps

    elif direction == 'LEFT':
        pos[0] -= steps

# Distance formula √(x² + y²)
distance = (pos[0]**2 + pos[1]**2) ** 0.5

print("The distance is", round(distance, 2))
````

### `Problem 13`:Write a program to print whether a given number is a prime number or not

````python
# Take input
num = int(input("Enter the number: "))

flag = True   # assume number is prime

# Check divisibility
for i in range(2, num // 2 + 1):
    if num % i == 0:
        flag = False
        break

# Print result
if flag:
    print("The number is prime")
else:
    print("The number is not prime")
````

### `Problem 14`:Print all the Armstrong numbers in a given range.
Range will be provided by the user<br>
Armstrong number is a number that is equal to the sum of cubes of its digits. For example 0, 1, 153, 370, 371 and 407 are the Armstrong numbers.

````python
# Take range input
n = int(input("Enter the range: "))

# Loop from 0 to n
for i in range(0, n + 1):

    num = i
    total = 0   # reset sum for each number

    # Find sum of cubes of digits
    while num > 0:
        lsb = num % 10
        total += lsb ** 3
        num = num // 10

    # Check Armstrong condition
    if total == i:
        print(i, end=" ")
````

### `Problem 15`:Calculate the angle between the hour hand and minute hand.

Note: There can be two angles between hands; we need to print a minimum of two. Also, we need to print the floor of the final result angle. For example, if the final angle is 10.61, we need to print 10.

Input:<br>
H = 9 , M = 0<br>
Output:<br>
90<br>
Explanation:<br>
The minimum angle between hour and minute
hand when the time is 9 is 90 degress.

````python
import math

# Take input
H = int(input("Enter Hours: "))
M = int(input("Enter Minutes: "))

# Convert to 12-hour format
H = H % 12

# Calculate angle between hands
angle = abs(30 * H - 5.5 * M)

# Take the smaller angle
min_angle = min(angle, 360 - angle)

# Print floor value
print(math.floor(min_angle))
````

### `Problem 17`:Sequence sum

1/1! + 2/2! + 3/3! + ............

````python
# Take input
n = int(input("Enter n: "))

result = 0   # to store final sum
fact = 1     # to store factorial

# Loop from 1 to n

for i in range(1, n + 1):
    # calculate i!
    fact = fact * i      

    # add term i/i!
    result = result + i / fact   

print(result)
````

### Q. write a program to find the prime numbers between the range provided by user

````python
# Take range input
lower = int(input("Enter the lower range: "))
upper = int(input("Enter the upper range: "))

# Loop through range
for i in range(lower, upper + 1):

    # Skip numbers <= 1
    if i <= 1:
        continue

    # Check for factors
    for j in range(2, i // 2 + 1):
        if i % j == 0:
            break
    else:
        # runs only if loop doesn't break
        print(i,end=" ")
````
