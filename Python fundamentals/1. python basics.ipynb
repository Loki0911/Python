### Q4:- Write a program to find the euclidean distance between two coordinates.Take both the coordinates from the user as input.

````python
#input the x and y value of two coordinate
x1,y1=float(input("Enter the x value of 1st coordinate")),float(input("Enter the y value of 1st coordinate"))
x2,y2=float(input("Enter the x value of 2nd coordinate")),float(input("Enter the x value of 2nd coordinate"))

#calculating the distance and printing it
distance=((y2-y1)**2+(x2-x1)**2)**(1/2)
print(f"the euclidean distance between ({x1},{y1}) and ({x2},{y2}) is {round(distance,2)}")
````

### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
# take number of legs and heads as input
heads=int(input("Enter the number of heads"))
legs=int(input("Enter the number of legs"))

# Using equations:
# d + c = heads
# 4d + 2c = legs

chickens= (4*heads-legs)//2
dogs=heads-chickens

print(chickens,"chikens and", dogs, "dogs")
````

### Q8:- Given the first 2 terms of an Arithmetic Series.Find the Nth term of the series. Assume all inputs are provided by the user.

````python
# input the first two term
a=int(input("Enter the first term"))
b=int(input("Enter the second term"))

#find the difference between two terms
d=b-a

#which term do user want ask to input
n=int(input("Enter the number of terms "))

#calculate the term using the formula and print it
term=a+(n-1)*d
print(f"the sum is {term}")
````

### Q10:- Given the height, width and breadth of a milk tank, you have to find out how many glasses of milk can be obtained? Assume all the inputs are provided by the user.



Input:<br>
Dimensions of the milk tank<br>
H = 20cm, L = 20cm, B = 20cm
<br><br>
Dimensions of the glass<br>
h = 3cm, r = 1cm

````python
# dimensions of tank
H=float(input("Enter the height of tank"))
L=float(input("Enter the length of tank"))
B=float(input("Enter the breadth of tank"))

#dimensions of glass
h=float(input("Enter the height of glass"))
r=float(input("Enter the radius of glass"))

#volume of milk tank
v1=L*B*H

#volume of glass
v2=2*3.1415*r*h

result=v1//v2

print(f"{result} glasses of milk can be obtained")
````
