###`Q1:` Join Tuples if similar initial element
While working with Python tuples, we can have a problem in which we need to perform concatenation of records from the similarity of initial element. This problem can have applications in data domains such as Data Science.

For eg.
```
Input  : test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]
Output : [(5, 6, 7, 8), (6, 10), (7, 13)]
```

````python
test_list = [(5, 6), (5, 7), (5, 8), (6, 10), (7, 13)]

unique = []

# Extract first elements from each tuple
for i in test_list:
    unique.append(i[0])

# Remove duplicates by converting to set, then back to list
unique = list(set(unique))

result = []

# Loop through each unique first element
for i in unique:
    result.append([i])   # Start a new sublist with the key
    
    # Find all tuples where first element matches current key
    for j in test_list:
        if j[0] == i:
            result[-1].append(j[-1])  # Append corresponding second values

# Convert inner lists to tuples
print(list(map(tuple, result)))
````

###`Q2:` Multiply Adjacent elements (both side) and take sum of right and lest side multiplication result.


For eg.
```
The original tuple : (1, 5, 7, 8, 10)
Resultant tuple after multiplication :

(1*5, 1*5+5*7, 7*5 + 7*8, 8*7 + 8*10, 10*8) -> (5, 40, 91, 136, 80)

output-(5, 40, 91, 136, 80)
```

````python
# Input tuple
mytuple = (1, 5, 7, 8, 10)

result = []

# Loop through each index of the tuple
for i in range(len(mytuple)):
    
    # Case 1: First element (no left neighbor)
    if i == 0:
        # Multiply current element with next element only
        result.append(mytuple[i] * mytuple[i + 1])
    
    # Case 2: Last element (no right neighbor)
    elif i == (len(mytuple)) - 1:
        # Multiply current element with previous element only
        result.append(mytuple[i] * mytuple[i - 1])

    # Case 3: Middle elements (have both neighbors)
    else:
        # Multiply current element with both neighbors and add results
        result.append(
            mytuple[i] * mytuple[i + 1] +  mytuple[i] * mytuple[i - 1]    
        )

# Convert result list to tuple
print(tuple(result))
````

###`Q3`: Check is tuples are same or not?
Two tuples would be same if both tuples have same element at same index
```
t1 = (1,2,3,0)
t2 = (0,1,2,3)

t1 and t2 are not same
```

````python
t1 = (1, 2, 3, 0)
t2 = (0, 1, 2, 3)

flag = False

if len(t1) == len(t2):
    for i, j in zip(t1, t2):
        # check if elements at same position match
        if i == j:
            # mark that a match exists
            flag = True
            continue
        else:
            # mismatch found → stop checking
            flag = False
            break

if flag:
    print("Tuples are same")
else:
    print("Tuples are not same")
````

###`Q4`: Count no of tuples, list and set from a list
```
list1 = [{'hi', 'bye'},{'Geeks', 'forGeeks'},('a', 'b'),['hi', 'bye'],['a', 'b']]

```
`Output:`

```
List-2
Set-2
Tuples-1
```

````python
list1 = [{'hi', 'bye'},{'Geeks', 'forGeeks'},('a', 'b'),['hi', 'bye'],['a', 'b']]

output = [0, 0, 0]

for i in list1:
    # check for list type
    if type(i) == list:
        output[0] += 1
    
    # check for tuple type
    elif type(i) == tuple:
        output[1] += 1
    
    # check for set type
    elif type(i) == set:
        output[2] += 1
    
    # ignore other types
    else:
        pass

print("Lists-", output[0], "\nTuples-", output[1], "\nSets-", output[2])
````

###`Q1:` Write a program to find set of common elements in three lists using sets.
```
Input : ar1 = [1, 5, 10, 20, 40, 80]
        ar2 = [6, 7, 20, 80, 100]
        ar3 = [3, 4, 15, 20, 30, 70, 80, 120]

Output : [80, 20]
```

````python
ar1 = [1, 5, 10, 20, 40, 80]
ar2 = [6, 7, 20, 80, 100]
ar3 = [3, 4, 15, 20, 30, 70, 80, 120]

# convert lists to sets
s1 = set(ar1)
s2 = set(ar2)
s3 = set(ar3)

# find common elements using intersection
print(list(s1 & s2 & s3))
````

### `Q1`: Key with maximum unique values

Given a dictionary with values list, extract key whose value has most unique values.

**Example 1:**

Input:

```bash
test_dict = {"CampusX" : [5, 7, 9, 4, 0], "is" : [6, 7, 4, 3, 3], "Best" : [9, 9, 6, 5, 5]}
```

Output:
```bash
CampusX
```

**Example 2:**

Input:
```bash
test_dict = {"CampusX" : [5, 7, 7, 7, 7], "is" : [6, 7, 7, 7], "Best" : [9, 9, 6, 5, 5]}
```

Output:
```bash
Best
```

````python
test_dict = {"CampusX": [5, 7, 9, 4, 0], "is": [6, 7, 4, 3, 3], "Best": [9, 9, 6, 5, 5]}

max_val = 0
max_key = ""

for keys, values in test_dict.items():
    # count unique elements using set
    if len(set(values)) > max_val:
        # update max unique count
        max_val = len(set(values))
        # store corresponding key
        max_key = keys

# print key with maximum unique values
print(max_key)
````

### `Q2`: Replace words from Dictionary. Given String, replace it’s words from lookup dictionary.

**Example 1:**

Input:

```bash
test_str = 'CampusX best for DS students.'
repl_dict = {"best" : "is the best channel", "DS" : "Data-Science"}
```

Output:

```bash
CampusX is the best channel for Data-Science students.
```

**Example 2:**

Input:
```bash
test_str = 'CampusX best for DS students.'
repl_dict = {"good" : "is the best channel", "ds" : "Data-Science"}
```

Output:
```bash
CampusX best for DS students.
```

````python
test_str = 'CampusX best for DS students.'
repl_dict = {"best": "is the best channel", "DS": "Data-Science"}

result = []

for i in test_str.split():
    # check if word exists in replacement dictionary
    if i in repl_dict:
        # replace with mapped value
        result.append(repl_dict[i])
    else:
        # keep original word
        result.append(i)

# join words back into string
print(' '.join(result))
````

### `Q3`: Convert List to List of dictionaries. Given list values and keys list, convert these values to key value pairs in form of list of dictionaries.

**Example 1:**

Input:
```bash
test_list = ["DataScience", 3, "is", 8]
key_list = ["name", "id"]
```

Output:

```bash
[{'name': 'DataScience', 'id': 3}, {'name': 'is', 'id': 8}]
```

**Example 2:**

Input:
```bash
test_list = ["CampusX", 10]
key_list = ["name", "id"]
```

Output:

```bash
[{'name': 'CampusX', 'id': 10}]
```

````python
test_list = ["DataScience", 3, "is", 8]
key_list = ["name", "id"]

output = []

# iterate in steps of 2 to pair elements
for i in range(0, len(test_list), 2):
    # create dictionary using keys and paired values
    output.append({
        key_list[0]: test_list[i],
        key_list[1]: test_list[i + 1]
    })

# print list of dictionaries
print(output)
````

### `Q4`: Convert a list of Tuples into Dictionary.

```bash
[("akash", 10), ("gaurav", 12), ("anand", 14), ("suraj", 20), ("akhil", 25), ("ashish", 30)]
```

Output:

```bash
{'akash': [10], 'gaurav': [12], 'anand': [14], 'suraj': [20], 'akhil': [25], 'ashish': [30]}
```

Input:
```bash
[('A', 1), ('B', 2), ('C', 3)]
```

### `Q5`: Sort Dictionary key and values List.

**Example 1:**

Input:

```bash
{'c': [3], 'b': [12, 10], 'a': [19, 4]}
```

Output:

```bash
{'a': [4, 19], 'b': [10, 12], 'c': [3]}
```

**Example 2:**

Input:

```bash
{'c': [10, 34, 3]}
```

Output:

```bash
{'c': [3, 10, 34]}
```

````python
d = {'c': [3], 'b': [12, 10], 'a': [19, 4]}

res = {}

# iterate over keys in sorted order
for i in sorted(d):
    # sort each list and assign to new dict
    res[i] = sorted(d[i])

print(res)
````
