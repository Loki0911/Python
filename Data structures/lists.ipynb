# List comprehension

##### 1. Add 1 to 10 numbers to a list

````python
# Create an empty list
l = []

# Loop from 1 to 10 (inclusive)
for i in range(1, 11):
    # Add each number to the list
    l.append(i)

# Print the final list
print(l)
````

````python
# Create a list using list comprehension
# It generates numbers from 1 to 10 and stores them in the list
l = [i for i in range(1, 11)]

# Print the list
print(l)
````

##### 2. scalar multiplication of vectors

````python
# Define a vector (list of numbers)
v = [2, 3, 4]

# Define a scalar value
s = -2

# Create an empty list to store the result
new_v = []

# Multiply each element of the vector by the scalar
for i in v:
    new_v.append(i * s)

# Print the new scaled vector
print(new_v)
````

````python
# Use list comprehension to multiply each element of the vector by the scalar
newV = [s * i for i in v]

# Print the new scaled vector
print(newV)
````

##### 6. Nested if with List Comprehensive
basket=['apple','guava','cherry','banana']
my_fruits=['apple','kiwi','grapes','banana']

###### add new list from my_fruits and items if the fruit exists in basket and also starts with 'a'

````python
# List of available fruits in the basket
basket = ['apple', 'guava', 'cherry', 'banana']

# List of fruits you have
my_fruits = ['apple', 'kiwi', 'grapes', 'banana']

# Create a new list with fruits that:
# 1. Exist in basket
# 2. Start with 'a'
bag = [fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]

# Print the final filtered list
print(bag)
````

##### 7.  print (3,3) matrix using list comprehension -> nested list conprehension

````python
# Create a nested list (2D list / matrix) using list comprehension
# Outer loop (j) controls rows
# Inner loop (i) generates values for each row

result = [[i * j for i in range(1, 4)] for j in range(1, 4)]

print(result)
````

l1=[1,2,3,4,]

````python
# First list
l1 = [1, 2, 3, 4]

# Second list
l2 = [5, 6, 7, 8]

# Use list comprehension to multiply each element of l1
# with each element of l2 (nested loop behavior)
l3 = [i * j for i in l1 for j in l2]

# Print the resulting list
print(l3)
````

# Zip function

````python
# First list
l1 = [1, 2, 3, 4]

# Second list
l2 = [-1, -2, -3, -4]

# Pair elements from both lists using zip
pairs = list(zip(l1, l2))
print(pairs)   # [(1, -1), (2, -2), (3, -3), (4, -4)]

# Add corresponding elements from both lists
result = [i + j for i, j in zip(l1, l2)]

# Print the result
print(result)   # [0, 0, 0, 0]
````

### `Problem 5:` You are given a list of integers. You are asked to make a list by running through elements of the list by adding all elements greater and itself.

i.e. Say given list is `[2,4,6,10,1]`
resultant list will be `[22,20,10,23]`.

For 1st element `2` ->> these are greater `(4+6+10)` values and `2` itself so on adding becomes `22`.

For 2nd element `4` ->> greater elements are `(6, 10)` and `4` itself, so on adding `20`

like wise for all other elememts.

`[2,4,6,10,1]`-->`[22,20,16,10,23]`




````python
l1 = [2,4,6,10,1]
l2 = [] 
sum = 0 

# iterate through each element in l1
for i in l1:
    # check each element again for comparison
    for j in l1:
        # if element is greater than or equal to current element add to the sim
        if j >= i:
            sum += j
    
    #append sum to list2
    l2.append(sum)

    #initialize sum to zero for next iteration
    sum = 0 

print(l2) 
````

### `Problem 6:` Find list of common unique items from two list. and show in increasing order

`Input`

```
num1 = [23,45,67,78,89,34]
num2 = [34,89,55,56,39,67]
```

`Output:`
```
[34, 67, 89]
```


````python
num1 = [23,45,67,78,89,34,34]  
num2 = [34,89,55,56,39,67]     

num3 = []  
# iterate through elements of num1
for i in num1:
    # check if element exists in num2 and is not already added
    if i in num2 and i not in num3:
        num3.append(i)

# sort the result in increasing order
num3.sort()

print(num3)
````

### `Problem 7:` Sort a list of alphanumeric strings based on product value of numeric character in it. If in any string there is no numeric character take it's product value as 1.

`Input:`
```
['1ac21', '23fg', '456', '098d','1','kls']
```

`Output:`
```
['456', '23fg', '1ac21', '1', 'kls', '098d']

```

````python
l1 = ['1ac21', '23fg', '456', '098d', '1', 'kls']

prod_val = [] 

# iterate through each string in the list
for item in l1:
    product = 1
    for char in item:

        # check if character is a digit multiply digit
        if char.isdigit():  
            product = product * int(char)  

    prod_val.append(product)
    
# zip products with original list, sort in descending order based on product
# then extract only the strings in sorted order
result = [i[1] for i in sorted(zip(prod_val, l1), reverse=True)]

print(result) 
````

### `Problem 10:` Add Space between Potential Words.

**Example:**

Input:

```bash
['campusxIs', 'bestFor', 'dataScientist']
```

Output:
```bash
['campusx Is', 'best For', 'data Scientist']
```

````python
l = ['campusxIs', 'bestFor', 'dataScientist']
result = []
# iterate through each word in the list
for items in l:
    # list to hold split parts
    temp = [[]]  

    # iterate through each character
    for char in items:
        # if uppercase, start a new word
        if char.isupper():  
            temp.append([])

        # add character to the current sublist
        temp[-1].append(char)  

    # to build the final string
    temp_string = ""

    # join each sublist into words and add space
    for i in temp:
        temp_string = temp_string + "".join(i) + " "

    # remove last extra space and store
    result.append(temp_string[0:-1])  

print(result)
````

### `Problem 11:` Write a program that can perform union operation on 2 lists

**Example:**

Input:

```bash
[1,2,3,4,5,1]
[2,3,5,7,8]
```

Output:
```bash
[1,2,3,4,5,7,8]
```

````python
l1 = [1,2,3,4,5,1]  
l2 = [2,3,5,7,8]    

result = [] 

# add unique elements from l1
for i in l1:
    if i not in result:
        result.append(i)

# add unique elements from l2
for j in l2:
    if j not in result:
        result.append(j)

 # sort the final list in ascending order
result.sort() 

print(result)
````

### `Problem 14:` Write a list comprehension that can transpose a given matrix

matrix = [<br>
  [1,2,3],<br>
          [4,5,6],<br>
          [7,8,9]<br>]<br>

[1, 4, 7]<br>
[2, 5, 8]<br>
[3, 6, 9]<br>

````python
# original matrix
matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]  

# transpose the matrix (convert rows to columns)
result = [[row[i] for row in matrix] for i in range(len(matrix))]

print(result)
````

### `Problem 15:` Write a list comprehension that can flatten a nested list

Input<br>
matrix = [<br>
  [1,2,3],<br>
          [4,5,6],<br>
          [7,8,9]<br>]<br>

Output:<br>
[1, 2, 3, 4, 5, 6, 7, 8, 9]

````python
# 2D list (matrix)
matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]  

# flatten the matrix into a single list
result = [item for row in matrix for item in row]

print(result)
````
