### `1.`  concept of class and objects

````python
class Atm:   # class

    # constructor -> runs automatically when object is created
    def __init__(self):

        print(id(self))   # memory address of current object

        # instance variables (object data)
        self.pin = ''
        self.balance = 0

    # method (behavior of object)
    def menu(self):

        user_input = input("""
        Hi how can I help you?
        1. Press 1 to create Pin
        2. Press 2 to change Pin
        3. Press 3 to check balance
        4. Press 4 to withdraw
        5. Anything else to exit
        """)

        if user_input == '1':
            self.create_pin()   # method calling

        elif user_input == '2':
            self.change_pin()

        elif user_input == '3':
            self.check_balance()

        elif user_input == '4':
            self.withdraw_cash()

        else:
            exit()   # function call required

    def create_pin(self):

        user_pin = input("Enter your pin: ")
        self.pin = user_pin   # updating instance variable

        user_balance = int(input('Enter the balance '))
        self.balance = user_balance

        print('Pin created successfully')
        self.menu()

    def change_pin(self):

        old_pin = input("Enter your current pin")

        if old_pin == self.pin:

            new_pin = input("Enter new pin")
            self.pin = new_pin

            print("Pin changed successfully")
            self.menu()

        else:
            print("Incorrect existing pin")
            self.menu()

    def check_balance(self):

        user_pin = input("Enter the pin")

        if user_pin == self.pin:

            print("Your current balance is {}".format(self.balance))
            self.menu()

        else:
            print("Incorrect pin")
            self.menu()

    def withdraw_cash(self):

        user_pin = input("Enter the pin")

        if user_pin == self.pin:

            cash = int(input('Enter the amount you want to withdraw'))

            # allows withdrawing complete balance also
            if cash <= self.balance:

                self.balance = self.balance - cash

                print("{} withdrawn successfully".format(cash))
                self.menu()

            else:
                print("Insufficient balance")
                self.withdraw_cash()

        else:
            print("Incorrect pin")
            self.menu()




````

### `2.` Making own datatype

````python
# creating custom datatype Fraction

class Fraction:      # Class

    # Parameterized constructor
    def __init__(self, x, y):

        # Instance variables
        self.num = x
        self.den = y

    # Magic method → controls object print format
    def __str__(self):

        return '{}/{}'.format(self.num, self.den)

    # Operator overloading for +
    def __add__(self, other):

        newnum = self.num * other.den + other.num * self.den
        newden = self.den * other.den

        return '{}/{}'.format(newnum, newden)

    # Operator overloading for -
    def __sub__(self, other):

        newnum = self.num * other.den - other.num * self.den
        newden = self.den * other.den

        return '{}/{}'.format(newnum, newden)

    # Operator overloading for *
    def __mul__(self, other):

        newnum = self.num * other.num
        newden = self.den * other.den

        return '{}/{}'.format(newnum, newden)

    # Operator overloading for /
    def __truediv__(self, other):

        newnum = self.num * other.den
        newden = self.den * other.num

        return '{}/{}'.format(newnum, newden)

    # User defined method
    def convert_to_decimal(self):

        return self.num / self.den







````

````python
# Object creation
fr1 = Fraction(3, 4)
fr2 = Fraction(1, 2)
````

````python
print(fr1 + fr2)
print(fr1 - fr2)
print(fr1 * fr2)
print(fr1 / fr2)
````

## `Q-1:` Rectangle Class
1. Write a Rectangle class in Python language, allowing you to build a rectangle with length and width attributes.

2. Create a Perimeter() method to calculate the perimeter of the rectangle and a Area() method to calculate the area of ​​the rectangle.

3. Create a method display() that display the length, width, perimeter and area of an object created using an instantiation on rectangle class.

Eg.
After making above classes and methods, on executing below code:-
```
my_rectangle = Rectangle(3 , 4)
my_rectangle.display()
```

`Output:`
```
The length of rectangle is:  3
The width of rectangle is:  4
The perimeter of rectangle is:  14
The area of rectangle is:  12
```


````python
class Rectangle:

    # Constructor method
    def __init__(self, l, b):
        self.len = l
        self.width = b
    
    # Method to calculate perimeter
    def perimeter(self):
        return 2 * (self.len + self.width)
    
    # Method to calculate area
    def area(self):
        return self.len * self.width
    
    # Method to display rectangle details
    def display(self):

        # Printing length
        print("The length of rectangle is:", self.len)

        # Printing width
        print("The width of rectangle is:", self.width)

        # Printing perimeter using perimeter() method
        print("The perimeter of rectangle is:", self.perimeter())

        # Printing area using area() method
        print("The area of rectangle is:", self.area())
````

````python
# Creating object of Rectangle class
rec = Rectangle(4,5)
````

````python
# Calling display method
rec.display()
````

## `Q-2: Bank Class`

1. Create a Python class called `BankAccount` which represents a bank account, having as attributes: `accountNumber` (numeric type), `name` (name of the account owner as string type), `balance`.
2. Create a constructor with parameters: `accountNumber, name, balance`.
3. Create a `Deposit()` method which manages the deposit actions.
4. Create a `Withdrawal()` method  which manages withdrawals actions.
5. Create an `bankFees()` method to apply the bank fees with a percentage of 5% of the balance account.
6. Create a `display()` method to display account details.
Give the complete code for the  BankAccount class.

Eg.
After making above classes and methods, on executing below code:-
```
newAccount = BankAccount(2178514584, "Mandy" , 2800)

newAccount.Withdrawal(700)

newAccount.Deposit(1000)

newAccount.display()
```

`Output:`
```
Account Number :  2178514584
Account Name :  Mandy
Account Balance :  3100 ₹
```

````python
class BankAccount:

    # Constructor method
    def __init__(self, acc_num, name, balance):
        self.acc_num = acc_num
        self.name = name
        self.balance = balance
    
    # Method to deposit money into account
    def Deposite(self, dep_amount):

        # Adding deposited amount to current balance
        self.balance = self.balance + dep_amount
        
    
    # Method to withdraw money from account
    def Withdrawal(self, withdrawal_amount):

        # Checking if balance is sufficient
        if withdrawal_amount > self.balance:
            print("Insufficient balance")

        else:
            # Deducting withdrawal amount
            self.balance = self.balance - withdrawal_amount

            # Applying bank fees after withdrawal
            self.bankFees()
    
    # Method to deduct 5% bank fee from current balance
    def bankFees(self):

        # Deducting 5% of current balance
        self.balance = self.balance - 0.05 * self.balance


    # Method to display account details
    def display(self):
       
        print("Account number :", self.acc_num)
        print("Account name :", self.name)
        print("Account balance :", self.balance, "₹")


# Creating object of BankAccount class
newAccount = BankAccount(2178514584, "Mandy", 2800)

# Withdrawing money
newAccount.Withdrawal(700)

# Depositing money
newAccount.Deposite(1000)

# Displaying final account details
newAccount.display()
````

## `Q-3:Computation class`

1. Create a `Computation` class with a default constructor (without parameters) allowing to perform various calculations on integers numbers.
2. Create a method called `Factorial()` which allows to calculate the factorial of an integer n. Integer n as parameter for this method

3. Create a method called `naturalSum()` allowing to calculate the sum of the first n integers 1 + 2 + 3 + .. + n. Integer n as parameter for this method.

4. Create a method called `testPrime()` in  the Calculation class to test the primality of a given integer n, n is Prime or Not? Integer n as parameter for this method.

5. Create  a method called `testPrims()` allowing to test if two numbers are prime between them. Two integers are prime to one another if they have only `1` as their common divisor. Eg. 4 and 9 are prime to each other.

5. Create a `tableMult()` method which creates and displays the multiplication table of a given integer. Then create an `allTablesMult()` method to display all the integer multiplication tables 1, 2, 3, ..., 9.

6. Create a static `listDiv()` method that gets all the divisors of a given integer on new list called  Ldiv. Create another `listDivPrim()` method that gets all the prime divisors of a given integer.

````python
class Computation:

    # Constructor method
    def __init__(self):
        pass
    

    # Method to calculate factorial of a number
    def Factorial(self, n):

        fact = 1

        # Multiplying numbers from 1 to n
        for i in range(1, n + 1):
            fact = fact * i
        
        return fact
    

    # Method to calculate sum of first n natural numbers
    def naturalSum(self, n):

        sum = 0
        str = ''

        # Adding numbers from 1 to n
        for i in range(1, n + 1):
            sum = sum + i

            # Storing 'i' in string (not used further)
            str = str + 'i'
        
        return sum
    

    # Method to check whether a number is prime or not
    def testPrime(self, n):

        # Checking divisibility from 2 to n-1
        for i in range(2, n):

            # If divisible, number is not prime
            if n % i == 0:
                return "Not prime"
        
        # If no divisor found, number is prime
        else:
            return "Prime"
        

    # Method to check whether two numbers are co-prime
    def testPrims(self, n1, n2):

        # Finding larger number
        a = n1 if n1 > n2 else n2

        # Checking common divisors
        for i in range(2, a):

            # If both numbers divisible by same number
            if n1 % i == 0 and n2 % i == 0:
                return "Not co-prims"
        
        # If no common divisor found
        else:
            return "co-Prims"
        

    # Method to print multiplication table of a number
    def tableMult(self, n):

        for i in range(1, 11):
            print(i * n)

    
    # Method to print multiplication tables from 1 to 10
    def alltableMults(self):

        for i in range(1, 11):

            print("\n", i, " : ", end=" ")

            for j in range(1, 11):
                print(i * j, end=' ')

        print('\n')


    # Method to list all divisors of a number
    def listDiv(self, n):

        lst = []

        # Checking divisors up to n//2
        for i in range(1, n // 2 + 1):

            if n % i == 0:
                lst.append(i)
    
        return lst


    # Method to list all prime divisors of a number
    def listDivPrim(self, n):

        lst = []
        
        # Finding divisors
        for i in range(1, n):

            if n % i == 0:

                # Ignoring 1 because it is not prime
                if i > 1:

                    prime = True

                    # Checking whether divisor is prime
                    for j in range(2, i):

                        if i % j == 0:
                            prime = False
                            break
                    
                    # Adding prime divisor to list
                    if prime:
                        lst.append(i)

        return lst


# Creating object of Computation class
intComp = Computation()

# Calling factorial method
print(intComp.Factorial(5))

# Calling natural sum method
print(intComp.naturalSum(5))

# Checking prime number
print(intComp.testPrime(5))

# Checking co-prime numbers
print(intComp.testPrims(4, 8))

# Printing multiplication table of 6
intComp.tableMult(6)

# Printing multiplication tables from 1 to 10
intComp.alltableMults()

# Printing divisors of 54
print(intComp.listDiv(54))

# Printing prime divisors of 54
print(intComp.listDivPrim(54))
````

## `Q-4`: Build flashcard using class in Python.

Build a flashcard using class in python. A flashcard is a card having information on both sides, which can be used as an aid in memoization. Flashcards usually have a question on one side and an answer on the other.

**Example 1:**

Approach:

- Create a class named FlashCard.
- Initialize dictionary fruits using __init__() method. Here you have to define fruit name as key and it's color as value. E.g., {"Banana": "yellow", "Strawberries": "pink"}
- Now randomly choose a pair from fruits by using _random_ module and store the key in variable _fruit_ and _value_ in variable color.
- Now prompt the user to answer the color of the randomly chosen fruit.
- If correct print correct else print wrong.

Output:
```bash
welcome to fruit quiz
What is the color of Strawberries
pink
Correct answer
Enter 0, if you want to play again: 0
What is the color of watermelon
green
Correct answer
Enter 0, if you want to play again: 1
```

````python
import random


class FlashCard:

    # Constructor method
    def __init__(self):

        self.__fruits = {
            'apple': 'red',
            'banana': 'yellow',
            'watermelon': 'green',
            'strawberry': 'pink',
            'guava': 'green'
        }


    # Method to start quiz
    def quiz(self):

        # Infinite loop to continue quiz until user exits
        while True:

            # Randomly selecting one fruit-color pair
            fruit, color = random.choices(list(self.__fruits.items()))[0]

            # Asking question to user
            print('What is the color of {}'.format(fruit))

            # Taking user input
            user_answer = input()

            # Checking whether answer is correct
            if user_answer.lower() == color:

                print('Correct answer')

            else:

                print("Wrong answer")


            # Asking whether user wants to continue
            option = int(input("Enter 0 if you want to play again: "))

            # Breaking loop if user enters anything other than 0
            if option != 0:
                break


# Welcome message
print("Welcome to the fruit quiz")


# Creating object of FlashCard class
fc = FlashCard()


# Calling quiz method
fc.quiz()
````

## `Q-5:` Problem 5 based on OOP Python.

TechWorld, a technology training center, wants to allocate courses for instructors. An instructor is identified by name, technology skills, experience and average feedback. An instructor is allocated a course, if he/she satisfies the below two conditions:
- eligibility criteria:
    - if experience is more than 3 years, average feedback should be 4.5 or more
    - if experience is 3 years or less, average feedback should be 4 or more
- he/she should posses the technology skill for the course

Identify the class name and attributes to represent instructors. Write a Python program to implement the class chosen with its attributes and methods.

**Note:**
- Consider all instance variables to be private and methods to be public.
- An instructor may have multiple technology skills, so consider instance variable, technology_skill to be a list.
- *check_eligibility()*: Return true if eligibility criteria is satisfied by the instructor. Else, return false
- *allocate_course(technology)*: Return true if the course which requires the given technology can be allocated to the instructor. Else, return false.

Represent a few objects of the class, initialize instance variables using setter methods, invoke
appropriate methods and test your program.

````python
class TechWorld:

    # Constructor method
    def __init__(self, name, tech_skills, experience, avg_feedback):

        self.__name = name
        self.__techskills = tech_skills
        self.__experience = experience
        self.__avgFeedback = avg_feedback
    

    # Method to check eligibility
    def check_eligibility(self):

    
        if self.__experience > 3 and self.__avgFeedback >= 4.5:
            return True
        

    
        elif self.__experience <= 3 and self.__avgFeedback >= 4:
            return True
        

        # Otherwise not eligible
        else:
            return False
        

    # Method to allocate course
    def allocate_course(self, technology):

        # Checking whether employee has required skill
        if technology in self.__techskills:

            # Checking eligibility
            if self.check_eligibility():
                return "{} is Eligible".format(self.__name)

            else:
                return "{} is not eligible".format(self.__name)
        

        # If technology skill not present
        else:
            return "{} doesn't know the skill".format(self.__name)
        

# Creating object
tech = TechWorld(
    "Lokesh",
    ["Data Science", "Python", "c"],
    3,
    3.9
)

# Calling method and printing result
print(tech.allocate_course("snake"))
````
