### 1. *args and **kwargs

````python
# concepts of *args

def multiply(*args):
    # initialize product to 1 (neutral value for multiplication)
    product = 1  

    # iterate through all arguments passed
    for i in args:
        # multiply each value with current product
        product = product * i  
    
    # show all input values (tuple of arguments)
    print(args)  
    
    # return final multiplication result
    return product  

multiply(1, 2, 3, 4, 5, 6)
````

````python
# Concept of **kwargs

def display(**kwargs):
    # iterate through key-value pairs in kwargs dictionary
    for (keys, values) in kwargs.items():
        # print each key and its corresponding value
        print(keys, '->', values)

# pass multiple keyword arguments (stored as a dictionary)
display(Nepal="Kathmandu", India="Delhi", Sirlanka="Colombu")
````

### Concept of local and global variable

````python
# Example 1

def g(y):
    # access global variable x and print it
    print(x)
    
    # use global x to compute and print x + 1
    print(x + 1)

# define global variable x
x = 5

# call function g and pass x (argument y is not used inside)
g(x)

# print global x again (unchanged)
print(x)
````

````python
# Example 2

def f(y):
    # create a local variable x inside the function
    x = 1
    
    # increment local x by 1
    x += 1
    
    # print the local value of x
    print(x)

# define global variable x
x = 5

# call function (y receives 5, but it's not used inside)
f(x)

# print global x (remains unchanged)
print(x)
````

### Higher order functions

````python
#example

# normal function → takes value and returns result
def square(x):
    return x**2

# higher-order function → takes another function as argument
def transform(f, l):
    # store results
    output = []

    # apply function f to each element
    for i in l:
        output.append(f(i))

    return output

l = [1, 2, 3, 4, 5]

# passing normal function to higher-order function
print(transform(square, l))

# passing lambda (anonymous function) to higher-order function
print(transform(lambda x: x**3, l))
````

### **`Problem-1:`** Write a Python function that takes a list and returns a new list with unique elements of the first list.

**Exercise 1:**

Input:

```bash
[1,2,3,3,3,3,4,5]
```

Output:

```bash
[1, 2, 3, 4, 5]
```

````python
def unique_items(l):
    # create an empty list to store unique elements
    unique=[]
    
    # iterate through each element in the input list
    for i in l:
        
        # check if the element is not already in the unique list
        if i not in unique:
            
            # add the element to the unique list
            unique.append(i)
    
    # return the list containing only unique elements
    return unique


unique_items([1,2,3,3,3,3,4,5])
````

### **`Problem 3:`** Write a Python function that accepts a string and calculate the number of upper case letters and lower case letters.

```
Sample String : 'CampusX is an Online Mentorship Program fOr EnginEering studentS.'
Expected Output :
No. of Upper case characters :  9
No. of Lower case Characters :  47
```

````python
s1 = 'CampusX is an Online Mentorship Program fOr EnginEering studentS.'

def upper_lower(s):
    # initialize counters
    lower_count = 0
    upper_count = 0

    # iterate through each character
    for i in s:

        # check for lowercase characters
        if i.islower():
            lower_count += 1
        
        # check for uppercase characters
        elif i.isupper():
            upper_count += 1

    # return counts (lower first, then upper)
    return lower_count, upper_count

# receive returned values correctly
lower, upper = upper_lower(s1)

# print results
print("No. of Upper case characters :", upper)
print("No. of Lower case characters:", lower)
````

### **`Problem 5:`** Write a Python function to check whether a number is perfect or not.

A Perfect number is a number that is half the sum of all of its positive divisors (including itself).

Example :

```
The first perfect number is 6, because 1, 2, and 3 are its proper positive divisors, and 1 + 2 + 3 = 6.
Equivalently, the number 6 is equal to half the sum of all its positive divisors: ( 1 + 2 + 3 + 6 ) / 2 = 6.

The next perfect number is 28 = 1 + 2 + 4 + 7 + 14. This is followed by the perfect numbers 496 and 8128.
```


````python
def is_perfect(num):

    # initialize sum of divisors
    sum=0

    # iterate from 1 to num//2 to find factors
    for i in range(1,num//2+1):
        
        # check if i is a divisor of num
        if num%i==0:
            
            # add divisor to sum
            sum+=i

    # check if sum of divisors equals the number
    if sum==num and (sum+num)/2 == num:
        
        # return if number is perfect
        return "perfect number"
    else:
        
        # return if number is not perfect
        return "Not a perfect number"
    
is_perfect(28)
````

### `Problem-7` Write a python function that accepts a string as input and returns the word with most occurence.

```
Input:
hello how are you i am fine thank you
```

```
Output
you -> 2
```

````python
def most_occured(s):
    # create an empty dictionary to store word counts
    d={}
    
    # iterate through each word in the string
    for i in s.split():
        
        # check if word is not in dictionary
        if i not in d:
            d[i]=1
        
        # increment count if word already exists
        else:
            d[i]+=1
    
    # find the maximum frequency value
    max_val=max(d.values())
        
    # find the word with maximum frequency
    for i in d:
        
        # check if current word has max frequency
        if d[i]==max_val:
            
            # return the word and its count
            return i,d[i]
        
s='hello how are you i am fine thank you'
x,y=most_occured(s)

print(x,'->',y)
````

### `Problem-8` Write a python function that receives a list of integers and prints out a histogram of bin size 10

```
Input:
[13,42,15,37,22,39,41,50]
```

```
Output:
{11-20:2,21-30:1,31-40:2,41-50:3}
```

````python
import math

l=[13,42,15,37,22,39,41,51]


def hist(l):
    # find the lower bound rounded down to nearest 10
    min_val=math.floor(min(l)/10)*10
    
    # find the upper bound rounded up to nearest 10
    max_val=math.ceil(max(l)/10)*10

    # create an empty dictionary to store histogram
    d={}

    # iterate through ranges of size 10
    for i in range(min_val,max_val,10):
        # initialize count for current range
        count=0
        
        # iterate through each element in list
        for j in l:

            # check if element falls within current range
            if i+1<=j<=i+10:
                count+=1
    
        # store range as key and count as value
        d[str(i+1)+'-'+str(i+10)]=count

    # return the histogram dictionary
    return d

result=hist(l)

print(result)
````

### `Problem-9` Write a python function that accepts a list of 2D co-ordinates and a query point, and then finds the the co-ordinate which is closest in terms of distance from the query point.

```
List of Coordinates
[(1,1),(2,2),(3,3),(4,4)]
Query Point
(0,0)
```

```
Output
Nearest to (0,0) is (1,1)
```

````python
def closest_distance(points,query):

    # create a list to store distances
    temp=[]
    
    # iterate through each point
    for i in points:
        
        # calculate squared Euclidean distance from query point
        distance= ((i[0]-query[0])**2+(i[1]-query[1])**2)
        
        # store distance in list
        temp.append(distance)

    # pair each index with its distance, sort by distance, and return closest point
    return  points[sorted(list(enumerate(temp)),key=lambda x:x[1])[0][0]]
    
points=[(1,1),(2,2),(3,3),(4,4)]
query=(4,1)

closest_distance(points,query)
````

### `Problem-10`:Write a python program that receives a list of strings and performs bag of word operation on those strings

https://en.wikipedia.org/wiki/Bag-of-words_model

````python
def bow(L):
    # create a set to store unique vocabulary
    vocab=set()

    # build vocabulary from all sentences
    for i in L:
        vocab.update(i.split())
        
    # initialize result list
    result=[]

    # iterate through each sentence
    for i in L:
        # add a new empty list for current sentence
        result.append([])
        
        # iterate through each word in vocabulary
        for j in vocab:
            
            # count occurrences of word in sentence and append
            result[-1].append(i.count(j))
    
    # print vocabulary set
    print(vocab)
    
    # return bag-of-words representation
    return result

L= [
    'cat mat rat cat',
    'sat bat fat cat rat',
    'pat cat mat rat'
]

bow(L)
````
