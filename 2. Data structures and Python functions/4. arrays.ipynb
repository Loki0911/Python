### `3.` Find Min/Max in a given array
```
L=[1,2,6,4,5]

```

````python
L=[1,2,6,4,5]

# initialize max and min with first element
max = L[0]
min = L[0]

# traverse list starting from second element
for i in range(1, len(L)):
    
    # update max if current element is greater
    if L[i] > max:
        max = L[i]

    # update min if current element is smaller
    if L[i] < min:
        min = L[i]

# print maximum and minimum values
print(max, min)
````

### `6.`  Rotate array to left d items
```
L= [1,2,3,4,5]
rotate = 2
```


````python
#solution 1

L= [1,2,3,4,5]
rotate = 2
# perform left rotation 'rotate' times
for i in range(rotate):

    # store first element
    temp = L[0]

    # shift all elements one position to the left
    for j in range(len(L) - 1):
        L[j] = L[j + 1]

    # place first element at the end
    L[-1] = temp

# print rotated list
print(L)
````

````python
#solution 2

L= [1,2,3,4,5]
rotate = 2

# slice the list from index 'rotate' to end (elements after rotation)
# and concatenate with elements from start to 'rotate'
result = L[rotate:] + L[:rotate]

# print the rotated list
print(result)
````

````python
#solution 3

L= [1,2,3,4,5]
rotate = 2

# perform left rotation 'rotate' times
for _ in range(rotate):

    # remove the first element from the list
    first = L.pop(0)

    # append the removed element to the end
    L.append(first)

# print the rotated list
print(L)
````

### `7.` Find intersection of 2 sorted arrays
```
a=[1,2,3,4,5,8]
b=[3,4,7,8]
```

````python
#solution 1

a=[1,2,3,4,5,8]
b=[3,4,7,8]

result=[]

# iterate through elements of first list
for i in a:

    # check if element is present in second list
    if i in b:
        result.append(i)

# print common elements (intersection)
print(result)
````

````python
#solution 2

a=[1,2,3,4,5,8]
b=[3,4,7,8]

result=[]

a_i = 0
b_j = 0

# use two pointers to traverse both sorted lists
while a_i < len(a) and b_j < len(b):

    # if elements are equal, add to result and move both pointers
    if a[a_i] == b[b_j]:
        result.append(a[a_i])
        a_i += 1
        b_j += 1

    # if element in 'a' is greater, move pointer of 'b'
    elif a[a_i] > b[b_j]:
        b_j += 1

    # if element in 'b' is greater, move pointer of 'a'
    else:
        a_i += 1

# print common elements (intersection)
print(result)

````

### `8.` Find continous subarray with a given sum(given non_negative numbers)
```
return the starting and ending index of the array
return 1st subarray in case of multiple
```

````python
#solution 1


L =[1,22,13,7,9,11,10]

result_sum=30

# iterate through each starting index of subarray
for i in range(len(L)):
    
    # temporary list to store current subarray
    sub_arr = []

    # extend subarray from current starting index
    for j in range(i, len(L)):

        # add current element to subarray
        sub_arr.append(L[j])

        # check if sum of current subarray matches target
        if sum(sub_arr) == result_sum:
            print(sub_arr)

````

````python
#solution 2

L =[1,22,13,7,9,11,10]

target= 27

# dictionary to store prefix sum and its index
d = {}

# variable to store running prefix sum
curr_sum = 0

# iterate through the array
for i in range(len(L)):

    # update prefix sum
    curr_sum = curr_sum + L[i]

    # check if subarray from index 0 to i matches target
    if curr_sum == target:
        print(0, i)

    # check if there exists a previous prefix sum such that
    # current_sum - previous_sum = target
    if (curr_sum - target) in d:
        print(d[curr_sum - target] + 1, i)

    # store prefix sum with index
    d[curr_sum] = i
````

### `9.` Find element with left side smaller/right side greater in an array
```
L= [3,1,2,5,8,7,9]
```

````python
#solution 1

L= [3,1,2,5,8,7,9]

# iterate through each element excluding first and last
for i in range(1, len(L) - 1):

    # assume current element is equilibrium element
    flag = True

    # check if all elements on left are smaller
    for j in range(0, i):
        if L[j] > L[i]:
            flag = False

    # check if all elements on right are greater
    for k in range(i, len(L)):
        if L[k] < L[i]:
            flag = False

    # if both conditions satisfied, print element
    if flag:
        print(L[i])
````

````python
#solution 2

L= [3,1,2,5,8,7,9]

# iterate through elements excluding first and last
for i in range(1, len(L) - 1):

    # check if current element is greater than all elements on left
    # and smaller than all elements on right
    if max(L[:i]) < L[i] < min(L[i+1:]):
        print(L[i])
````

````python
#solution 3

L= [3,1,2,5,8,7,9]
# list to store prefix maximums (max element from left to i)
max_arr = []

# initialize first maximum as first element
max_val = L[0]

# compute prefix maximum array
for i in L:

    # update max if current element is larger
    if i > max_val:
        max_val = i

    # store current maximum
    max_arr.append(max_val)

print(max_arr)


# list to store suffix minimums (min element from i to end)
min_arr = []

# initialize last minimum as last element
min_val = L[-1]

# compute suffix minimum array from right to left
for i in range(len(L) - 1, -1, -1):

    # update min if current element is smaller
    if L[i] < min_val:
        min_val = L[i]

    # insert at front to maintain order
    min_arr.insert(0, min_val)

print(min_arr)


# find element which is greater than all left elements
# and smaller than all right elements
for i in range(1, len(L) - 1):
    if max_arr[i - 1] < L[i] < min_arr[i + 1]:
        print(L[i])
````

### `10.` Maximum sum subarray
```
L=[-2,4,7,-1,6,-11,14,3,-1,-6]
```

````python
#solution 1

L=[-2,4,7,-1,6,-11,14,3,-1,-6]
# dictionary to store sum of each subarray as key and subarray as value
d = {}

# iterate over all possible starting points of subarray
for i in range(0, len(L)):

    # temporary list to build subarray
    subarray = []

    # extend subarray from index i to end
    for j in range(i, len(L)):

        # add current element to subarray
        subarray.append(L[j])

        # store sum of current subarray in dictionary
        d[sum(subarray)] = subarray

# find maximum subarray sum stored in dictionary keys
max_val = max(d.keys())

# print the subarray corresponding to maximum sum
for i in d:
    if i == max_val:
        print(d[i])
````

````python
# solution 2:  Kadane's Algorithm: Maximum Sum Subarray with tracking sequence

# input list
L = [-2, 4, 7, -1, 6, -11, 14, 3, -1, -6]

# current running sum
curr_sum = 0

# current subarray sequence
curr_seq = []

# best (maximum) sum found so far
best_sum = L[0]

# best subarray corresponding to best sum
best_seq = []

# iterate through each element
for i in L:

    # decide whether to extend previous subarray or start new one
    if curr_sum + i > i:
        curr_sum = curr_sum + i
        curr_seq.append(i)
    else:
        curr_sum = i
        curr_seq = [i]

    # update best result if current sum is better
    if curr_sum > best_sum:
        best_sum = curr_sum
        best_seq = curr_seq.copy()

# print maximum sum subarray
print(best_seq)
````

### `11.` maximum product sub array

````python
# array input
arr = [-2, 6, -3, -10, 0, 2]

# length of array
n = len(arr)

# max product ending at current index
currMax = arr[0]

# min product ending at current index (important for negative numbers)
currMin = arr[0]

# overall maximum product found so far
maxProd = arr[0]

# iterate from second element
for i in range(1, n):

    # store previous values before updating
    tempMax = currMax
    tempMin = currMin

    # compute maximum product ending at current index
    currMax = max(arr[i], arr[i] * tempMax, arr[i] * tempMin)

    # compute minimum product ending at current index
    currMin = min(arr[i], arr[i] * tempMax, arr[i] * tempMin)

    # update global maximum product
    maxProd = max(maxProd, currMax)

# print final maximum product subarray value
print(maxProd)
````

### `13.` move all -ve numbers to the end

````python
arr = [-2, 6, -3, -10, 0, 2]

# pointer j keeps track of position for next non-negative number
j = 0

# iterate through the array
for i in range(len(arr)):

    # if current element is non-negative
    if arr[i] >= 0:

        # swap current element with element at position j
        arr[i], arr[j] = arr[j], arr[i]

        # move j forward
        j += 1

# print rearranged array (non-negatives first, negatives later)
print(arr)
````

### `14. ` find the pythagorus triplets in an array

````python
# list of numbers
arr = [3, 1, 4, 6, 5]

n = len(arr)

# square all elements (Pythagorean check works on squares)
for i in range(n):
    arr[i] = arr[i] * arr[i]

# sort squared values
arr.sort()

# fix the largest element (c^2)
for c in range(n - 1, 1, -1):

    i = 0
    j = c - 1

    # two-pointer search for a^2 + b^2 = c^2
    while i < j:

        # if found a valid triplet
        if arr[i] + arr[j] == arr[c]:
            print(int(arr[i]**0.5), int(arr[j]**0.5), int(arr[c]**0.5))
            i += 1
            j -= 1

        # if sum is smaller, increase left pointer
        elif arr[i] + arr[j] < arr[c]:
            i += 1

        # if sum is larger, decrease right pointer
        else:
            j -= 1
````
