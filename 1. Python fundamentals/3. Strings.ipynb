### 1. Find the length of a given string without using len() function

````python
# Take input from the user
s = input("Enter the string: ")

# Initialize a counter to store length
count = 0

# Loop through each character in the string
for i in s:
    count += 1   # Increase count by 1 for each character

# Print the result using formatted string
print("The length of string '{}' is {}".format(s, count))
````

#### 3. Count the frequency of a particular character in a provided string.
#### eg: 'hello how are you' is the string. The frequency of h in this string is 2

````python
# Take string input from user
s = input("Enter the string: ")

# Take the character to count
c = input("Enter the character: ")

# Initialize counter
count = 0

# Loop through each character in the string
for i in s:
    # Check if current character matches the given character
    if i == c:
        count += 1   # Increase count if matched

# Display the result
print(f"The frequency of character '{c}' in string '{s}' is {count}")
````

### 4. Write a program which can remove a particular character from a string

````python
# Take string input from user
s = input("Enter the string: ")

# Take the character to remove
c = input("Enter the character to remove: ")

# Initialize an empty string to store result
newstr = ''

# Loop through each character in the string
for i in s:
    # Add character to new string only if it is not equal to c
    if i != c:
        newstr += i

# Print the updated string
print(newstr)
````

### 5. WAP that can check whether a given string is palindrome or not

````python
# Take input string from user
s = input("Enter a string: ")

# Assume the string is a palindrome initially
flag = True

# Loop only till half of the string
for i in range(len(s) // 2):
    # Compare characters from start and end
    if s[i] != s[len(s) - i - 1]:
        flag = False
        print("Entered string is not a palindrome")
        break

# If no mismatch found
if flag:
    print("Entered string is a palindrome")
````

### 6. Write a program to count the number of words in a string without split()

````python
# Take input string from user
s = input("Enter a string: ")

# Initialize word count and flag
count = 0
flag = False   # Tracks whether we are inside a word

# Loop through each character
for i in s:
    # If character is not space and we are not already in a word
    if i != ' ' and not flag:
        count += 1      # New word found
        flag = True     # Now inside a word

    # If character is a space, mark that we are outside a word
    elif i == ' ':
        flag = False

# Print result
print("Total words in '{}' is {}".format(s, count))
````

### 7. WAP to convert a string to title case without using the title()

````python
# Take input string from user
s = input("Enter a string: ")

# Split the string into words
for i in s.split():
    # Convert first letter to uppercase and keep rest as it is
    print(i[0].upper() + i[1:], end=' ')
````

### 8. WAP to convert an integer to string

````python
# Take integer input from user
num = int(input("Enter the integer value: "))

# String containing all digits
digits = '0123456789'

# Initialize empty string (avoid using 'str' as variable name)
result = ''

# Handle special case when number is 0
if num == 0:
    result = '0'

# Convert integer to string manually
while num > 0:
    result = digits[num % 10] + result   # get last digit and add to front
    num = num // 10                      # remove last digit

# Print result and its type
print(result, type(result))
````

### ``Problem 13``:Given string contains a combination of the lower and upper case letters. Write a program to arrange the characters of a string so that all lowercase letters should come first.

Given:

`str1 = PyNaTive`

Expected Output:

`yaivePNT`

````python
# Take input string from user
s = input("Enter the string")

# Initialize empty strings to store lowercase and uppercase characters
low = ''
up = ''

# Loop through each character in the string
for i in s:
    # If character is lowercase, add to 'low'
    if i.islower():
        low = low + i
    # If character is uppercase, add to 'up'
    elif i.isupper():
        up = up + i

# Print lowercase characters first, then uppercase characters
print(low + up)
````

### `Problem 14:`Take a alphanumeric string input and print the sum and average of the digits that appear in the string, ignoring all other characters.

Input:

`hel123O4every093`

Output:

```
Sum: 22
Avg: 2.75
```

````python
# Take input from user
s = input("Enter the alphanumeric string: ")

sum = 0
count = 0

# Loop through each character
for i in s:
    if i.isdigit():
        sum = sum + int(i)
        count += 1

# Avoid division by zero
if count > 0:
    avg = sum / count
    print("Input string:", s)
    print("Total number of digits found:", count)
    print("Sum of all digits:", sum)
    print("Average of digits:", avg)
else:
    print("No digits found in the string.")
````

### `Problem 16`: Check whether the string is Symmetrical.

**Statement:** Given a string. the task is to check if the string is symmetrical or not. A string is said to be symmetrical if both the halves of the string are the same.

**Example 1:**

`Input`

```bash
khokho
```

`Output`

```bash
The entered string is symmetrical
```

````python
# Take input string
s = input("Enter the string: ")

flag = False

# Check if length is even
if len(s) % 2 == 0:
    # Compare first half with second half
    if s[0:len(s)//2] == s[len(s)//2:]:
        flag = True
else:
    # For odd length, ignore the middle character
    if s[0:len(s)//2] == s[len(s)//2 + 1:]:
        flag = True

# Print result
if flag:
    print("The string is symmetrical.")
else:
    print("The string is not symmetrical.")
````

### `Problem 18`: Find uncommon words from two Strings.

**Statement:** Given two sentences as strings **A** and **B**. The task is to return a list of all uncommon words. A word is uncommon if it appears exactly once in any one of the sentences, and does not appear in the other sentence. Note: A sentence is a string of space-separated words. Each word consists only of lowercase letters.

**Example 1:**

`Input:`

```bash
A = "apple banana mango"
B = "banana fruits mango"
```

`Output:`

```bash
['apple', 'fruits']
```

````python
# Take two sentences as input
A = input("Enter the first sentence: ")
B = input("Enter the second sentence: ")

# Initialize empty list to store uncommon words
L = []

# Check words in A that are not in B
for i in A.split(" "):
    if i not in B.split(" ") and i not in L:
        L.append(i)

# Check words in B that are not in A
for j in B.split(" "):
    if j not in A.split(" ") and j not in L:
        L.append(j)

# Print the list of uncommon words
print("Uncommon words between the two sentences:", L)
````

### `Problem 19`: Word location in String.

**Statement:** Find a location of a word in a given sentence.

**Example 1:**

`Input:`

```bash
Sentence: We can go to swimming in sunny days.

word: sunny
```

`Output:`

```bash
Location of the word is 7.
```

Note- Don't use index/find functions

````python
# Take input sentence and word
s = input("Enter the sentence: ")
w = input("Enter the word: ")

pos = 0
found = False

# Loop through words with position
for i in s.split(" "):
    pos += 1
    if i == w:
        found = True
        break

# Print result
if found:
    print("'{}' in '{}' is at position {}".format(w, s, pos))
else:
    print("'{}' is not found in '{}'".format(w, s))
````

### `Problem 20`: Write a program that can remove all the duplicate characters from a string. User will provide the input.

````python
# Take input string
s = input("Enter the string: ")

# Initialize empty string to store result
result = ''

# Loop through each character
for i in s:
    # Add character only if it is not already in result
    if i not in result:
        result = result + i

# Print final result
print("After removing all duplicate characters from '{}' the result is '{}'".format(s, result))
````
